```python
from ultralytics import YOLO

# Load pre-trained YOLO model or your custom model if available
model = YOLO('yolov8n.pt')  # Replace with your trained model path if necessary
```

```python
model.train(data='/content/sample_data/data.yaml', epochs=50, imgsz=640)
```

```python
# Perform inference
results = model('/content/sample_data/test/images/Brain-Tumor-Overview_mp4-0129_jpg.rf.60e041d6464a8825ad6d5dd5c669cbaa.jpg')

# Access the first result in the results list
result = results[0]  # YOLOv8 returns a list of results, even for a single image

# Display the result
result.show()

# Save the result to disk
result.save('/content/results/image.jpg')  # Specify a valid filename with an extension


```
